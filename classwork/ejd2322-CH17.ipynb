```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 


steps= [.01, .02, .03,.05, .1, .2, .3, .5]
for h in steps:
    
    k = 1.0        
    x_start = 0    
    x_end = 10     
    y_0 = 1.0      
    v_0 = 0.0      

    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0


    for i in range(1, len(x_values)):
  
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
        
  
    plt.plot(x_values, y_values, label=f'Position y(x) with h={h}')
    plt.plot(x_values, v_values, label=f'Velocity v(x) with h={h}', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method (varying values of h)')
plt.grid(True)
plt.legend(fontsize = 4.5, loc = 'lower left')
plt.figure(figsize=(10, 5))
plt.show()
```

```python
##The solution using Euler's Method becamme unstable at around h=0.5 resulting in the position and velocity crazy proportions and large oscillations compared to the rest of the step size.
##Yes,the solution becomes more and more unstable as the step size increases, showing that instability is growing.
```

```python

play_with_k= [1,2,3,4,5]
 
for k in play_with_k:
    h= 0.1
    x_start = 0   
    x_end = 10     
    y_0 = 1.0     
    v_0 = 0.0      
    
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    
    y_values[0] = y_0
    v_values[0] = v_0

   
    for i in range(1, len(x_values)):
        
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
       
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    
    plt.plot(x_values, y_values, label=f'Position y(x) with h={h} and k={k}')
    plt.plot(x_values, v_values, label=f'Velocity v(x) with h={h} and k={k}', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method (varying values of h)')
plt.grid(True)
plt.legend(fontsize = 4.5, loc = 'lower left')
plt.figure(figsize=(10, 5))
plt.show()
```

```python
## As the k value increases the oscillations grow bigger and the wave function continuously reaches new peaks. The bigger k value the bigger the wave. The lower the k value the more concentrated the wave function is around the 0 line
```
