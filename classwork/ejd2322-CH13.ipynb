```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline

# Matrix Multiplication
def custom_matrix_multiplication(A, B):
    n = len(A)
    C = [[0]*n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

# Numpy Matrix Multiplication
def numpy_matrix_multiplication(A, B):
    return np.dot(A, B)
```

```python
# Random Matrices
matrix_sizes = [3]
for _ in range(7):
    matrix_sizes.append(matrix_sizes[-1]*2)
```

```python
custom_times = []
numpy_times = []

# Compare runtime of methods
for size in matrix_sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)
    
    # Custom matrix multiplication time
    start_time = time.time()
    custom_matrix_multiplication(A, B)
    custom_time = time.time() - start_time
    custom_times.append(custom_time)
    
    # Numpy matrix multiplication time
    start_time = time.time()
    numpy_matrix_multiplication(A, B)
    numpy_time = time.time() - start_time
    numpy_times.append(numpy_time)
    
    print(f"Matrix size: {size}x{size}")
    print(f"Custome time: {custom_time:.6f} seconds")
    print(f"Numpy time: {numpy_time:.6f} seconds")
```

```python
# Plot times
plt.figure(figsize=(10, 6))
plt.plot(matrix_sizes, custom_times, label='Custom Matrix Multiplication', marker='o')
plt.plot(matrix_sizes, numpy_times, label='Numpy Matrix Multiplication', marker='x')
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('Time (seconds)')
plt.title('Matrix Multiplication: Custom vs Numpy')
plt.legend()
plt.grid(True)
plt.show()
```
