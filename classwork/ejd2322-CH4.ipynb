```python
import numpy as np

piEstimate = 0.0

for k in range(5000):
    term = (-1)**k / (2*k + 1)
    piEstimate += term

    currentPi = 4 * piEstimate

    
    if (k + 1) % 100 == 0:
        relativeError = abs(currentPi - np.pi) / np.pi
        print(f"Terms: {k+1:4d} | π estimate: {currentPi:.10f} | Relative error: {relativeError:.6e}")
```
