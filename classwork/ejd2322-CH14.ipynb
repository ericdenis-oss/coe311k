```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.sin(x)

# Interval and discretization
a, b = 0, 2 * np.pi
N = 100

x = np.linspace(a, b, N)
h = (b - a) / (N - 1)

# True derivative
true_derivative = np.cos(x)

# Initialize arrays
forward_diff = np.zeros(N)
backward_diff = np.zeros(N)
central_diff = np.zeros(N)

# Forward Difference
for i in range(N - 1):
    forward_diff[i] = (f(x[i + 1]) - f(x[i])) / h
forward_diff[-1] = np.nan  # undefined at last point

# Backward Difference
for i in range(1, N):
    backward_diff[i] = (f(x[i]) - f(x[i - 1])) / h
backward_diff[0] = np.nan  # undefined at first point

# Central Difference
for i in range(1, N - 1):
    central_diff[i] = (f(x[i + 1]) - f(x[i - 1])) / (2 * h)
central_diff[0] = np.nan
central_diff[-1] = np.nan

# Plotting (4 subplots)
plt.figure(figsize=(12, 8))

# True derivative
plt.subplot(2, 2, 1)
plt.plot(x, true_derivative, label="cos(x)", color='black')
plt.title("True Derivative")
plt.legend()
plt.grid()

# Forward Difference
plt.subplot(2, 2, 2)
plt.plot(x, forward_diff, label="Forward Difference", color='blue')
plt.title("Forward Difference")
plt.legend()
plt.grid()

# Backward Difference
plt.subplot(2, 2, 3)
plt.plot(x, backward_diff, label="Backward Difference", color='red')
plt.title("Backward Difference")
plt.legend()
plt.grid()

# Central Difference
plt.subplot(2, 2, 4)
plt.plot(x, central_diff, label="Central Difference", color='green')
plt.title("Central Difference")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
