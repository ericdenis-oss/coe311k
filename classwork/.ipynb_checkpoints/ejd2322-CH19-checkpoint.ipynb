```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

h_list=[.1,.01,.001]
for h in h_list:
    x_start, x_end = 0, 30
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps+1)
    y_values = np.zeros(num_steps + 1)
    
    y_values[0] = 0 
    
    def f(x, y):
        return -1000*y+3000-2000*np.exp(-x)
    
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h *f(x_values[i], y_values[i])
        
    plt.plot(x_values, y_values, label=f"Euler's Method Approximation, h size: {h}")
    print(f"h={h}, max y={np.max(y_values)}, min y={np.min(y_values)}")
        

x_exact = np.linspace(0, 30, 10000)
exact_sol= 3-0.998*np.exp(-1000*x_exact)-2.002*np.exp(-x_exact)

plt.plot(x_exact,exact_sol, "k--", label="Exact Solution")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -1000*y+3000-2000*np.exp(-t) using Euler's Method")
plt.ylim(0,10)
plt.legend()
plt.show()
```
