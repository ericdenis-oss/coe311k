```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x),
x, np.sin(2 * x))
plt.show()
```

```python
plt.plot(x, np.sin(x), 'r-o',
x, np.cos(x), 'g--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 500)

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'r')
plt.title("sin(x)")

plt.subplot(2, 2, 2)
plt.plot(x, np.tan(x), 'g')
plt.title("tan(x)")

plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x))


plt.subplot(2, 2, 4)
plt.hist(np.tan(x), bins=20)

plt.show()
```
