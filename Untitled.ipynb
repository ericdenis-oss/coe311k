# U.S. GDP Growth Rate Analysis Using Curve Fitting & Interpolation

## Introduction

This project analyzes U.S. quarterly GDP growth rates from 2010 to 2023 using three numerical methods: least squares linear regression, polynomial fitting, and cubic spline interpolation. The dataset, sourced from the U.S. Bureau of Economic Analysis (BEA), includes both stable economic periods and extreme fluctuations during the COVID-19 pandemic.

The goal is to implement these methods from scratch, evaluate their performance using RMSE, residuals, and plots, and determine which method is best suited for interpolation versus trend analysis.

### Part A — Cubic Spline Interpolation


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# -----------------------
# DATA (integer indices)
# -----------------------
y = np.array([1.7, 0.1, 2.3, 2.7, 1.7, 5.0, 3.0, 1.5, 3.5, 1.8,
              1.3, 2.5, 3.1, -5.1, -28.1, 33.8, 6.3, -1.6, 2.4, 3.3])
x = np.arange(1, len(y)+1)

# -----------------------
# Build tridiagonal system
# -----------------------
def setup_tridiagonal(x, y):
    n = len(x)
    h = np.diff(x)

    A = np.zeros(n-2)
    B = np.zeros(n-2)
    C = np.zeros(n-2)
    D = np.zeros(n-2)

    for i in range(1, n-1):
        A[i-1] = h[i-1]
        B[i-1] = 2 * (h[i-1] + h[i])
        C[i-1] = h[i]

        D[i-1] = 6 * ((y[i+1]-y[i])/h[i] - (y[i]-y[i-1])/h[i-1])

    return A, B, C, D

# -----------------------
# Thomas Algorithm
# -----------------------
def thomas(a, b, c, d):
    n = len(b)

    for i in range(1, n):
        w = a[i-1] / b[i-1]
        b[i] -= w * c[i-1]
        d[i] -= w * d[i-1]

    x = np.zeros(n)
    x[-1] = d[-1] / b[-1]

    for i in range(n-2, -1, -1):
        x[i] = (d[i] - c[i]*x[i+1]) / b[i]

    return x

# Solve for second derivatives
A, B, C, D = setup_tridiagonal(x, y)
M_internal = thomas(A.copy(), B.copy(), C.copy(), D.copy())

M = np.zeros(len(x))
M[1:-1] = M_internal

# Evaluate spline
def eval_spline(x, y, M, x_eval):
    for i in range(len(x)-1):
        if x[i] <= x_eval <= x[i+1]:
            h = x[i+1] - x[i]
            a = (x[i+1] - x_eval) / h
            b = (x_eval - x[i]) / h

            return (a*y[i] + b*y[i+1] +
                    ((a**3 - a)*M[i] + (b**3 - b)*M[i+1])*(h**2)/6)

# Plot spline
x_smooth = np.linspace(1, len(x), 200)
y_smooth = np.array([eval_spline(x, y, M, xi) for xi in x_smooth])

plt.figure()
plt.scatter(x, y, label="Data Points")
plt.plot(x_smooth, y_smooth, label="Cubic Spline")

plt.xlabel("Quarter Index")
plt.ylabel("GDP Growth Rate (%)")
plt.title("Cubic Spline Interpolation of U.S. GDP Growth")
plt.legend()
plt.grid()

plt.show()
```

#### 2. Discussion

The spline is smooth and continuous up to the second derivative, and it passes exactly through all data points. However, near the COVID-19 quarters, the spline bends sharply due to extreme values.

#### 3. Runge Phenomenon & Outliers

High-degree polynomials can oscillate significantly (Runge phenomenon), but cubic splines reduce this by using piecewise functions. However, splines are still sensitive to large outliers, which cause sharp curvature locally.

Better alternatives include smoothing splines or weighted least squares.


### Part B — Polynomial and Least Squares Comparison




```python
# Vandermonde matrix
def vandermonde(x, degree):
    V = np.zeros((len(x), degree+1))
    for i in range(len(x)):
        for j in range(degree+1):
            V[i,j] = x[i]**j
    return V

degree = 4
V = vandermonde(x, degree)

# Normal equation
coeffs = np.linalg.solve(V.T @ V, V.T @ y)

def poly(x_val, coeffs):
    return sum(coeffs[i]*x_val**i for i in range(len(coeffs)))

y_poly = np.array([poly(xi, coeffs) for xi in x_smooth])

plt.figure()
plt.scatter(x, y, label="Data Points")
plt.plot(x_smooth, y_smooth, label="Cubic Spline")
plt.plot(x_smooth, y_poly, label="Degree 4 Polynomial")

plt.xlabel("Quarter Index")
plt.ylabel("GDP Growth Rate (%)")
plt.title("Cubic Spline vs Polynomial Fit")
plt.legend()
plt.grid()

plt.show()
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

y_poly_data = np.array([poly(xi, coeffs) for xi in x])
y_spline_data = np.array([eval_spline(x, y, M, xi) for xi in x])

print("Spline RMSE:", rmse(y, y_spline_data))
print("Polynomial RMSE:", rmse(y, y_poly_data))

# Residual plot
residuals = y - y_poly_data
plt.figure()
plt.scatter(x, residuals)

plt.axhline(0)
plt.xlabel("Quarter Index")
plt.ylabel("Residual (Actual - Predicted)")
plt.title("Residuals of Polynomial Fit")
plt.grid()

plt.show()
```

```python
mask = (x < 14) | (x > 17)
x_lin = x[mask]
y_lin = y[mask]

n = len(x_lin)

m = (n*np.sum(x_lin*y_lin) - np.sum(x_lin)*np.sum(y_lin)) / \
    (n*np.sum(x_lin**2) - (np.sum(x_lin))**2)

b = (np.sum(y_lin) - m*np.sum(x_lin)) / n

print("Slope:", m)

plt.figure()
plt.scatter(x_lin, y_lin, label="Data (No COVID)")
plt.plot(x, m*x + b, label="Linear Fit")

plt.xlabel("Quarter Index")
plt.ylabel("GDP Growth Rate (%)")
plt.title("Least Squares Linear Fit (Excluding COVID Period)")
plt.legend()
plt.grid()

plt.show()
```

## Conclusion

Cubic splines provide the best interpolation due to accuracy and smoothness. Polynomial fitting captures overall trends but misses local variations, while linear regression is useful only for long-term approximations. The presence of extreme events like COVID-19 highlights the importance of selecting appropriate numerical methods based on the problem context.
