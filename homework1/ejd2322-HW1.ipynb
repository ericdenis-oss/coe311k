```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

#import libraries 





def MonteCarloPi(samples):

    #sets intial for samples inside Circle to 0
    
    insideCircle = 0
    
    #for number of samples generate a random x & y between 0 and 1 and determine if the distance is within a circle radius
    #of 1 if it is add to counter (inside Circle) divide by num of samples and multiply by 4 to account for whole circle
    
    for i in range(samples):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        
        if x**2 + y**2 <= 1:
            insideCircle += 1
            
    return 4 * insideCircle / samples
```

```python
sampleSize = list(range(10,10000,20))
errors = []

#for sample size of 10 to 10000 in increments of 20 input into function and the output will be the pi estimate,
#subtract the estimate with the actual value of pi and add to error list for each sample 

for n in sampleSize:
    piEstimate = MonteCarloPi(n)
    error = abs(piEstimate - math.pi)
    errors.append(error)

    
```

```python
#plots error 

plt.figure()
plt.plot(sampleSize, errors)
plt.xlabel("# of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Approximation Error")
plt.grid(True)
plt.show()
```

```python
runs = 100
n_samples = 1000
true_pi = np.pi

abs_errors = []
rel_errors = []

for _ in range(runs):
 
    piEstimate = MonteCarloPi(n_samples)
    
    
    abs_err = abs(true_pi - piEstimate)
    abs_errors.append(abs_err)
    
    
    rel_err = abs((true_pi - piEstimate) / true_pi)
    rel_errors.append(rel_err)


plt.figure(figsize=(10, 5))

plt.plot(abs_errors, label='Absolute Error', color='blue')
plt.plot(rel_errors, label='Relative Error', color='red', linestyle='--')

plt.title('Monte Carlo PI: Error Analysis (100 Runs @ 1000 Samples)')
plt.xlabel('Run Number')
plt.ylabel('Error')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
