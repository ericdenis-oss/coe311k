```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

def MonteCarloPi(samples):

    insideCircle = 0
    
    for i in range(samples):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        
        if x**2 + y**2 <= 1:
            insideCircle += 1
            
    return 4 * insideCircle / samples
```
